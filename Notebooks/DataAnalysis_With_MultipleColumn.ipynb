```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("../Data/googleplaystore.csv")
df
```

```python
df.isnull().sum()
```

```python
impute = SimpleImputer(missing_values=np.nan,strategy="mean")
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
df = df.dropna()
```

### Q1. How many free apps are there in category ART_AND_DESIGN

```python
df
```

```python
## Convert dataframe to array
data = df.values
```

```python
### Using Loops with array
c = 0
for i in data:
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Free'):
        c += 1
print(f"There are {c} free apps in ART_AND_DESIGN category")
    
```

```python
## Using Pandas with dataframe
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
print(f"There are {len(df_pr[df_pr['Type'] == 'Free'])} free apps in ART_AND_DESIGN category")
```

```python
### Using Loops with array
c = 0
for i in data:
    if(i[1] == 'ART_AND_DESIGN' and i[6] == 'Paid'):
        c += 1
print(f"There are {c} Paid apps in ART_AND_DESIGN category")
```

```python
## Using Pandas with dataframe
print("There are", len(df_pr[df_pr['Type']=='Paid']), "free apps in ART_AND_DESIGN category")
```

### How many apps are there in ART_AND_DESIGN with rating greater than 4.5

```python
### Using Loops with array
c = 0
for i in data:
    if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5):
        c += 1
print(f"There are {c} apps in ART_AND_DESIGN category with rating greater than 4.5")
```

```python
## Using Pandas with dataframe
print(f"There are {len(df_pr[df_pr['Rating'] > 4.5])} apps in ART_AND_DESIGN category with rating greater than 4.5")
```

### How many apps are there in FAMILY with rating greater than 4.5 and free

```python
### Using Loops with array
c = 0
for i in data:
    if(i[1] == 'FAMILY' and i[2] > 4.5 and i[6] == 'Free'):
        c += 1
print(f"There are {c} apps in FAMILY category with rating greater than 4.5 and free")
```

```python
## Using Pandas with dataframe
df_fa = df[df['Rating'] > 4.5]
df_fa = df_fa[df_fa['Category'] == 'FAMILY']
print("There are ",len(df_fa[df_fa['Type'] == 'Free'])," free apps in FAMILY category with rating greater than 4.5")
```

### List all the free apps with rating more than 4.5 and category as ART_AND_DESIGN 

```python
### Using Loops with array
print("-"*90)
print("These are the free apps where category is ART_AND_DESIGN and rating is greater then 4.5")
print("-"*90)
for i in data:
    if(i[1] == 'ART_AND_DESIGN' and i[2] > 4.5 and i[6] == 'Free'):
        print(i[0])
```

```python
## Using Pandas with dataframe
print("-"*90)
print("These are the free apps where category is ART_AND_DESIGN and rating is greater then 4.5")
print("-"*90)
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Type'] == 'Free']
df_pr = df_pr[df_pr['Rating'] > 4.5].values ## convert to array to use loops
for i in df_pr:
    print(i[0])
```
