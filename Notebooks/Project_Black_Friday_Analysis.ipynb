```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (12, 6), title = 'Purchase distribution by Age')
```

```python
for i in df['Age'].unique():
    print(i)
```

```python
for i in df['Age'].unique():
    print(i, " : ", len(df[df['Age'] == i]))
```

```python
df.columns
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i, df[df['Age'] == i]['Product_ID'].nunique()])
```

```python
data = pd.DataFrame(lst, columns = ['Age', 'Products'])
```

```python
data
```

```python
data.plot.bar(x = 'Age', figsize = (8, 6))
```

```python
df.groupby('Age')['Purchase'].sum().plot(kind = 'bar', figsize = (12, 6), title = 'Amount distribution by Age')
```

```python
df.groupby('Age')['Purchase'].mean().plot(kind = 'pie', figsize = (6, 6), autopct = '%.2f')
```
