```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../Data/BlackFriday.csv')
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = 'Gender', data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years',hue = 'Marital_Status', data = df)
```

```python
sns.countplot(x = 'City_Category', hue = 'Age', data = df)
```

# Plot pie chart

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'pie', figsize = (6,6), autopct = '%0.1f')
```

```python
df.groupby('Stay_In_Current_City_Years')['Purchase'].mean().plot(kind = 'pie', figsize = (6,6), autopct = '%0.1f')
```

### Which occupation category has the highest spending?

```python
df.groupby('Occupation')['Purchase'].sum().sort_values()
```

```python
df.groupby('Occupation')['Purchase'].sum().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Occupation')['Purchase'].mean().sort_values().plot(kind = 'bar')
```

### How does spending vary across City Categories A, B, and C

```python
df.groupby('City_Category')['Purchase'].sum().sort_values().plot(kind = 'bar')
```

```python
sns.countplot(x = 'City_Category', data = df)
```

### How many unique products are ordered by each occupation?

```python
df.groupby('Occupation')['Product_ID'].nunique().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_Category_1')['Product_ID'].size().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_Category_1')['Purchase'].sum().sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_ID')['Purchase'].sum().nlargest(10).sort_values().plot(kind = 'bar')
```

```python
df.groupby('Product_ID')['Purchase'].sum().nlargest(10).sort_values().plot(kind = 'bar')
```
