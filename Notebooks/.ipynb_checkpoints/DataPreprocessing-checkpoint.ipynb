# Categorical Data

### Total number of catogeries

```python
df['Category'].unique()
```

```python
print(f"There are total {len(df['Category'].unique())} catogeries")
```

```python
for i in df['Category'].unique():
    print(i)
```

```python
a = 0
for i in df['Category']:
    if i == "ART_AND_DESIGN":
        a += 1
print(f"There are {a} applications of category ART_AND_DESIGN")
```

### Number of applications in each category

```python
categories = {}

for name in df['Category'].unique():
    ct = 0
    for i in  df['Category']:
        if i == name:
            ct += 1
    categories[name] = ct
categories
```

### Total numbers apps in each content rating

```python
Content_rating_count = {}
for name in df['Content Rating'].unique():
    ct = 0
    for i in  df['Content Rating']:
        if i == name:
            ct += 1
    Content_rating_count[name] = ct
Content_rating_count
```

### Total number of App - Types. Free and Paid

```python
df['Type'].unique()
```

```python
f = 0
p = 0
for i in df['Type']:
    if i == 'Free':
        f += 1
    elif i == 'Paid':
        p += 1
print(f"There are {f} free apps and {p} paid apps in the google playstore")
```

### Percentage of free apps in the dataset

```python
print(f"{int(f/(f + p) * 100)} % applications are free.")
```

```python
print(f"{int(p/(f + p) * 100)} % applications are paid.")
```

### Replacing null values with mean, median, mode values

```python
data.isnull().sum()
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(data.iloc[:,2:3].values)
data.iloc[:,2:3] = imputer.transform(data.iloc[:,2:3].values)
data
```
