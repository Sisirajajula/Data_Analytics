```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("../Data/googleplaystore.csv")
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3] = impute.transform(df.iloc[:,2:3].values)
df = df.dropna()
```

### Q1. Name the apps that are in ART_AND_DESIGN category with rating more than 4.5 in ascending order with respective ratings.

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
df_pr = df_pr[df_pr['Rating'] > 4.5]
df_pr.sort_values(by = 'Rating', ascending = True)
```

### Q2. Which category is having maximum average rating.

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending = False)
```

### Q3. How many apps are there in each category in descending order.

```python
df.groupby('Category')['Type'].count().sort_values(ascending = False)
```
