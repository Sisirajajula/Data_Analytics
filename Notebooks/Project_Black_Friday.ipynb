```python
df = pd.DataFrame({'Ratio': [len(data[data['Gender'] == 'M']), len(data[data['Gender'] == 'F'])]}, index = ['Male', 'Female'])
df.plot.pie(y='Ratio', figsize = (6,6), autopct = '%.2f')
```

## Another method using groupby
